```python

#1.GAUSS (concave graph)

'''Gausian Array
numpy.meshgrid()â€“ It is used to create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing. 
'''

x, y = np.meshgrid(np.linspace(-2,2,20), np.linspace(-2,2,20))
dist = np.sqrt(x*x+y*y)

sigma = 1 
muu = 0.000

'''
Gauss :: e^(-(dist-mmu)**2/(2*sigma**2))
'''
gauss = np.exp(-( (dist-muu)**2 / ( 2.0 * sigma**2 )))
```

```python
import matplotlib.pyplot as plt

plt.plot(gauss)
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-2,2,15), np.linspace(-2,2,15))
dst = np.sqrt(x*x+y*y)
  
# Intializing sigma and muu
sigma = 1
muu = 0.000
  
# Calculating Gaussian array
gauss = np.exp(-( (dst-muu)**2 / ( 2.0 * sigma**2 ) ) )

plt.plot(gauss)
plt.show()
```

```python
import scipy as sc
from scipy import special
import numpy as np
from numpy import r_

def drumhead_height(n,k,dist,angle,t):
    kth_zero = special.jn_zeros(n,k)[-1]
    return (np.cos(t) * np.cos(n*angle) * special.jn(n,dist*kth_zero))

theta = np.r_[0:2*np.pi:50j]
radius = np.r_[0:1:50j]
x = np.array([r * np.cos(theta) for r in radius])
y = np.array([r * np.sin(theta) for r in radius])
z = np.array([drumhead_height(1, 1, r, theta, 0.5) for r in radius])

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig, rect=(0, 0.05, 0.95, 0.95))
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='RdBu_r', vmin=-0.5, vmax=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.set_xticks(np.arange(-1, 1.1, 0.5))
ax.set_yticks(np.arange(-1, 1.1, 0.5))

ax.set_zlabel('Z')
plt.show()
```
